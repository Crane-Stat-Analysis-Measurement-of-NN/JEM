```python
import matplotlib.pyplot as plt
```

```python
exp='debug2'
```

```python
res = {'epoch': [],'loss': [],'train': [], 'valid': [], 'test': []}
fig = plt.figure()
ax = plt.gca()


for var, color in [('train','green'),('test','blue'),('valid','red')]:
    file = f'./{exp}/track_{var}.csv'

    res['epoch']=[]
    res['loss']=[]
    with open(file,'r') as rf:
        for line in rf.readlines()[1:]:
            e, l, a = (float(i.strip()) for i in line.split(','))
            res['epoch'].append(e)
            res['loss'].append(l)
            res[var].append(a)
        ax.plot(res['epoch'],res[var],color,label=var+' acc')
        if var=='train':
            ax.plot(res['epoch'],res['loss'],'black',label="loss",linestyle='dashed')


h,l = ax.get_legend_handles_labels()
handles = [h[0], h[2], h[3],h[1]]
labels = [l[0], l[2], l[3],l[1]]
plt.legend(handles, labels)
fig.show()
```
