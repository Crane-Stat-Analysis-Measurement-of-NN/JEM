```python
import matplotlib.pyplot as plt
import matplotlib.patches as pat
import csv
import pandas as pd
```

```python
class pac(object):pass

def pac_df(df):
    p=pac()
    p.Epoch = range(0,len(df))
    p.Loss = df['Loss']
    p.Acc = df['Acc']
    return p
```

```python
def get_pac(filename):
    df = pd.read_csv(filename)
    df=df.drop(range(0,50),axis=0)
    return pac_df(df)
```

```python
def plotE(*args,**kwargs):
    ax=plt.gca()
    ax.add_patch(pat.Ellipse(*args,**kwargs))
```

```python
def thisplot(p,title,y='Loss',color='black',x='Epoch'):
    plt.plot(getattr(p,x),getattr(p,y),label=y,color=color)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(title)
    plt.show()
```

```python
dir='test_X-ent_augmented/'
p = get_pac(dir+'validation.csv')
thisplot(p,'Validation','Loss','red')
thisplot(p,'Validation','Acc','blue')

p=get_pac(dir+'test.csv')
thisplot(p,'Test','Loss','red')
thisplot(p,'Test','Acc','blue')

p=get_pac(dir+'train.csv')
thisplot(p,'Train','Loss','red')
thisplot(p,'Train','Acc','blue')
```
